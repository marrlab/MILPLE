```python
import os
import pickle
from matplotlib import pyplot as plt
import numpy as np
from sklearn.metrics import auc
```

```python
def calculate_auc(data):
    patients = list(data.keys())
    x = range(100)
    a = 0
    for p in patients:
        y = data[p]
        a += auc(x,y)
    return a/len(patients)
```

```python
def calculate(what):
    with open("mds_insertion_"+what+".dat", "rb") as f:
        ins = pickle.load(f)
    y = 100-calculate_auc(ins)
    with open("mds_deletion_"+what+".dat", "rb") as f:
        dele = pickle.load(f)
    x = calculate_auc(dele)
    return x/100, y/100
```

```python
import matplotlib
font = {'family' : 'Arial',
        'weight' : 'normal',
        'size'   : 14}

matplotlib.rc('font', **font)
plt.figure(dpi=300)
x, y =calculate("gradcam")
plt.plot(x,y, "o",label="GradCAM" )
x, y =calculate("lrp")
plt.plot(x,y, "o",label="LRP" )
x, y =calculate("input_iba")
plt.plot(x,y, "o",label="Input IBA" )
x, y =calculate("feat_iba")
plt.plot(x,y, "o",label="IBA" )
# x, y =calculate("rnd")
# plt.plot(x,y, "o",label="RND" )
# # plt.xlim([0, 100])
# plt.ylim([0, 100])

# plt.ylabel("Insertions AUC")
# plt.xlabel("Deletions AUC")
# plt.legend(loc="lower right")
plt.savefig("insdel.svg", format="svg")
```

```python
def calculate2(what):
    with open("insertion_"+what+".dat", "rb") as f:
        ins = pickle.load(f)
    a = [ins[x] for x in ins]
    
    with open("deletion_"+what+".dat", "rb") as f:
        dele = pickle.load(f)
    b = [dele[x] for x in ins]
    
    return np.sum(np.array(a),axis=0)/len(np.sum(np.array(a),axis=0)),np.sum(np.array(b),axis=0)/len(np.sum(np.array(b),axis=0))
```
